**WARNING**: This notebook handles write operations, queue resets, schema expansions, and task insertions. Execute cells with caution.

```python
import os
import json
import pandas as pd
from sqlalchemy import create_engine, text

# 1. Connection & Setup
DATABASE_URL = os.environ.get("DATABASE_URL", "postgresql://localhost/postgres")
if DATABASE_URL.startswith("postgres://"):
    DATABASE_URL = DATABASE_URL.replace("postgres://", "postgresql://", 1)

engine = create_engine(DATABASE_URL)

def sql_write(query_str, params=None):
    """Executes transactional write/mutation queries."""
    with engine.begin() as conn:
        return conn.execute(text(query_str), params or {})

print("Admin tools session initialized. Write capabilities active.")
```

### Queue Management & Task Insertion
Use this section to queue new scraper commands or cancel hung jobs.

```python
# Example: Queue a fresh script execution task safely using parameterized JSON
task_payload = json.dumps({"command": "python3 scraper/scrape_redfin.py"})

insert_query = """
    INSERT INTO ai_tasks (task_type, status, priority, payload) 
    VALUES ('execute_script', 'pending', 1, :payload);
"""

sql_write(insert_query, {"payload": task_payload})
print("Successfully queued new scraper task via admin panel.")
```
