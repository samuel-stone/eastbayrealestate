# 02 | Lead Analytics & Property Insights
Analyze scraped real estate property leads, examine price-per-square-foot distributions, and visualize market metrics.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine, text

# 1. Connection & Setup
DATABASE_URL = os.environ.get("DATABASE_URL", "postgresql://localhost/postgres")
if DATABASE_URL.startswith("postgres://"):
    DATABASE_URL = DATABASE_URL.replace("postgres://", "postgresql://", 1)

engine = create_engine(DATABASE_URL)

def sql(query_str, params=None):
    with engine.connect() as conn:
        return pd.read_sql(text(query_str), conn, params=params)

print("Lead analytics environment initialized.")
```

### Property Price Distribution Chart
Visualize the distribution of asking prices across current valid property leads.

```python
# Pull valid leads with price data
df_leads = sql("SELECT address, price, beds, baths, sqft FROM leads_sandbox WHERE price IS NOT NULL;")

if not df_leads.empty:
    plt.figure(figsize=(10, 5))
    plt.hist(df_leads['price'].dropna() / 1e6, bins=15, color='#2b5c8f', edgecolor='black', alpha=0.8)
    plt.title('Property Price Distribution ($ Millions)', fontsize=12, fontweight='bold')
    plt.xlabel('Asking Price ($M)', fontsize=10)
    plt.ylabel('Count', fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("No price data currently available in leads_sandbox to chart.")
```
