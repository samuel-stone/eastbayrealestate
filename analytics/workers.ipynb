# 03 | Worker Task Monitoring & Queue Health
Monitor asynchronous background tasks, inspect queue status (`pending`, `in_progress`, `completed`, `failed`), and troubleshoot scraper executions.

```python
import os
import pandas as pd
from sqlalchemy import create_engine, text

# 1. Connection & Setup
DATABASE_URL = os.environ.get("DATABASE_URL", "postgresql://localhost/postgres")
if DATABASE_URL.startswith("postgres://"):
    DATABASE_URL = DATABASE_URL.replace("postgres://", "postgresql://", 1)

engine = create_engine(DATABASE_URL)

def sql(query_str, params=None):
    with engine.connect() as conn:
        return pd.read_sql(text(query_str), conn, params=params)

print("Worker monitoring environment ready.")
```

### Task Queue Status Breakdown
Check the breakdown of tasks by status to ensure the worker loop is processing correctly.

```python
queue_status_query = """
    SELECT status, COUNT(*) AS count 
    FROM ai_tasks 
    GROUP BY status 
    ORDER BY count DESC;
"""
df_status = sql(queue_status_query)
df_status
```
