```
import pandas as pd
pg8000 = 'postgresql://user:password@host:port/dbname'
conn = pg8000.connect()
cur = conn.cursor()
```

```
import pandas as pd
leads_df = pd.read_sql_query('SELECT * FROM leads', conn)
prospect_features_df = pd.read_sql_query('SELECT * FROM prospect_features', conn)
```

```
import pandas as pd
leads_df['date'] = pd.to_datetime(leads_df['date'])
prospect_features_df['date'] = pd.to_datetime(prospect_features_df['date'])
```

```
import pandas as pd
leads_df['age'] = leads_df['date'].dt.year - leads_df['birth_date'].dt.year
prospect_features_df['age'] = prospect_features_df['date'].dt.year - prospect_features_df['birth_date'].dt.year
```

```
import plotly.express as px
fig = px.histogram(leads_df, x='age', title='Age Distribution')
```
