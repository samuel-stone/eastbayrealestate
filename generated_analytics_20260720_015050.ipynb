```python
import os
import pandas as pd
import psycopg2
import plotly.express as px

# Establish secure connection to production database via environment variables
db_url = os.environ.get('DATABASE_URL')
conn = psycopg2.connect(db_url)
query = '''
    SELECT l.address, l.city, l.status, 
           COALESCE(f.building_permit_count_24m, 0) as building_permit_count_24m,
           COALESCE(f.project_count, 0) as project_count
    FROM leads l
    LEFT JOIN prospect_features f ON l.id = f.lead_id
    ORDER BY building_permit_count_24m DESC
    LIMIT 200
'''
df = pd.read_sql(query, conn)
conn.close()

print(f'Loaded {len(df)} records for analysis.')
df.head()
```
